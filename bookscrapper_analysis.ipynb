### Data Explorations

```python
# Explore data values values
df.groupby(by="category").agg({
    "title": "count",
    "price": ["min", "max"],
    "stars": "mean"
}).sort_values(("title", "count"), ascending=False)
```

```python
# Adding rating metrics
df["rating"] = df.stars.apply(lambda x: "Excellent" if x > 3 else("Bad" if x < 3 else "Good"))
df.groupby(by="rating").upc.nunique().sort_values(ascending=False)
```
