### Import Data

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Refrence docs (https://matplotlib.org/3.5.3/api/_as_gen/matplotlib.pyplot.html)
```

### Basic Graph

```python
x = [0,1,2,3,4]
y = [0,3,6,9,12]

# Line 2
x2 = np.arange(0,4.5,0.5)

# Resize your Graph (dpi specifies pixels per inch. When saving probably should use 300 if possible)
plt.figure(figsize = (8,5), dpi = 80)

# plt.plot(x,y, label = "2x", color = "red", linewidth = 2, marker = "x", linestyle = "-", markersize = 10, markeredgecolor = "blue")
# fmt = "[color][marker][line]"
plt.plot(x2, x2**2, "k", label = "x^2")

# Plot part of the graph as line
plt.plot(x[:4], y[:4], "rx-", label = "x^2")

# Plot remainder of graph as a dot
plt.plot(x[3:], y[3:], "r--")

# Add a title
plt.title("My First Graph", fontdict = {"fontname": "Calibri", "fontsize": 20})

# X and Y labels
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# X, Y axis tickmarks (scale of your graph)
plt.xticks([0,1,2,3,4])
# plt.yticks([0,2,4,6,8,10])

# Add a legend
plt.legend()

# Save figure (dpi 300 is good when saving so graph has high resolution)
# plt.savefig("mygraph.png", dpi = 300)

# Show plot
plt.show()
```

### Bar Chart

```python
labels = ["A", "B", "C", "D", "E", "F", "G", "H"]
values = [1,4,2,5,3,2,7,9]

plt.figure(figsize = (8,5), dpi = 80)

# Add a title
plt.title("My Bar Chart", fontdict = {"fontname": "Calibri", "fontsize": 20})

colors = ["#F4B084", "#A9D08E", "#9BC2E6", "#FFD966"]

bars = plt.bar(labels, values, width = 0.3, color = colors)

plt.xlabel("Label")
plt.ylabel("Value")

plt.legend

# plt.savefig("barchart.png", dpi = 300)

plt.show()
```

### Load Fifa Data

```python
fifa = pd.read_csv("./matplot_data/fifa_data.csv")

fifa.head(5)
```

### Histogram

```python
# i = 40
# bins = []

# while i < 100:
#     bins.append(i)
#     i+=2

bins = [(40+2*i) for i in range(0,30)]

plt.figure(figsize = (8,5), dpi = 80)

plt.hist(fifa.Overall, bins = bins, color = "#9BC2E6")

# p = 250
# players = []
# while p <= 2500:
#     players.append(p)
#     p+=250

players = [(250+250*j) for j in range(0,10)]

plt.xticks(bins)
plt.yticks(players)

plt.title("Distribution of Player Skills in FIFA 2018", fontdict = {"fontname": "Calibri", "fontsize": 20})
plt.ylabel("Number of Players")
plt.xlabel("Skill Level")
plt.grid(visible = True, which = "both", axis = "both")

# plt.savefig("histogram.png", dpi = 300)

plt.show()
```

### Pie Chart

```python
plt.figure(figsize = (8,5), dpi = 80)

plt.style.use("ggplot")

fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[fifa.Weight >= 200].count()[0]

weights = [light,light_medium, medium, medium_heavy, heavy]
label = ["Under 125", "125-150", "150-175", "175-200", "Over 200"]

plt.title("  Weight of Professional Football Players (lbs)", fontdict = {"fontname": "Calibri", "fontsize": 20})

explode = [0.2,0.01,0.01,0.01,0.2]

plt.pie(weights, labels = label, explode = explode, pctdistance = .6,autopct = '%.2f%%')
plt.show()
```

### Box and Whiskers Chart

```python
plt.figure(figsize = (5,8), dpi = 80)

plt.style.use("default")

barcelona = fifa.loc[fifa.Club == "FC Barcelona"]["Overall"]
madrid = fifa.loc[fifa.Club == "Real Madrid"]["Overall"]
revs = fifa.loc[fifa.Club == "New England Revolution"]["Overall"]

labels = ["FC Barcelona","Real Madrid","NE Revolution"]

# bp = plt.boxplot([barcelona, madrid, revs], labels=["a","b","c"], boxprops=dict(facecolor="red"))
bp = plt.boxplot([barcelona, madrid, revs], labels = labels, patch_artist = True)

plt.title("Professional Soccer Team Comparison", fontdict = {"fontname": "Calibri", "fontsize": 20})

plt.ylabel("FIFA Overall Rating")

for box in bp["boxes"]:
    # change outline color
    box.set(color="#000000", linewidth = 1)
    # change fill color
    box.set(facecolor = "#F4B084" )
    # change hatch
    box.set(hatch = "/")
    
plt.show()
```
