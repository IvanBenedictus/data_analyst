```python
import pandas as pd
```

### Import All Datasets

```python
package_df = pd.read_csv("./tourist_data/tourism_package.csv")
package_df
```

```python
place_df = pd.read_csv("./tourist_data/tourism_place.csv")
place_df = place_df.drop(["Description", "Coordinate", "Unnamed: 11", "Unnamed: 12"], axis=1)
place_df
```

```python
rating_df = pd.read_csv("./tourist_data/tourism_rating.csv")
rating_df
```

```python
user_df = pd.read_csv("./tourist_data/user_city.csv")
user_df
```

### Check Duplicates

### Check Null Values

```python
# Checking for missing values
check_column = ["Package", "City", "Place_Tourism1", "Place_Tourism2", "Place_Tourism3"]
package_df[package_df[check_column].isnull().any(axis=1)]
```

```python
# Checking for missing values
check_column = [x for x in range(9) if x != 6]
place_df[place_df.iloc[:, check_column].isnull().any(axis=1)]
```

```python
# Checking for missing values
rating_df[rating_df.isnull().any(axis=1)]
```

```python
# Checking for missing values
user_df[user_df.isnull().any(axis=1)]
```

### Save Modified File
